```python
# removing digits , rating , optional space from society attribute and converting to lowercase
import re

df['society'] = df['society'].apply(
    lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', name).strip().lower() if pd.notnull(name) else name
)
```

```python
df['price'].value_counts()
```

```python
# making the price column into the same unit

df=df[(df['price']!='Price on Request') & (df['price']!='price')]
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
df.head()
```

```python
#price_per_sqft
def clean_price(price):
    try:
        if isinstance(price, str):
            cleaned = price.split('/')[0]              # ₹ 5,000/sq.ft. → ₹ 5,000
            cleaned = cleaned.replace('₹', '')         # remove ₹
            cleaned = cleaned.replace(',', '')         # remove comma
            return float(cleaned.strip())              # convert to float
    except:
        return None
    return None

df['price_per_sqft'] = df['price_per_sqft'].apply(clean_price)
```
